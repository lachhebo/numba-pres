```python
test_array = list(range(1000000))
```

```python
from numba import jit, njit, vectorize
import random
```

```python
def octo_tag(talent_skoolers):
    output = []
    
    for skooler in talent_skoolers:
        if skooler % 100 > 95:
            output.append('Mleng')
        elif skooler % 100 > 75:
            output.append('Duc')
        else:
            output.append(0)
        
    return output

octo_tag_jit = njit(octo_tag)
```

```python
octo_tag(test_array)
```

```python
octo_tag_jit(test_array)
```
